### **Daily Coding Problem: Problem #57 [Medium]** - **Amazon**

```
Given a string s and an integer k, break up the string into multiple lines such that each line has a length of k or less. You must break it up so that words don't break across lines. Each line has to have the maximum possible amount of words. If there's no way to break the text up, then return null.

You can assume that there are no spaces at the ends of the string and that there is exactly one space between each word.

For example, given the string "the quick brown fox jumps over the lazy dog" and k = 10, you should return: ["the quick", "brown fox", "jumps over", "the lazy", "dog"]. No string in the list has a length of more than 10.

```

````python
def break_string_recursive_helper(
    words, line_width, cost_list, word_index
):
  
    # base line cost
    base_cost = 0

    # base case
    if word_index >= len(words): 
        return base_cost

    # checking memo
    if cost_list[word_index] is not None:
        return cost_list[word_index]

    # loop exit condition
    word_fits = True
    words_traversed = False

    # word index traversal
    next_index = word_index + 1

    # cost of current word index taken only
    # next words that meet loop condition
    words_cost_guess = []

    # line lenght for cost evaluation
    line_len = len(words[word_index])

    # check if word itself fits
    if not (line_len <= line_width):
        word_fits = False
    
    while word_fits and not words_traversed:
        
        # next words minimum cost calculation
        next_word_cost = break_string_recursive_helper(
            words, line_width, cost_list, next_index
        )

        # word index cost evaluation
        line_cost_with_next_word = pow(line_width - line_len, 3)
        
        # current total cost
        total_cost_taken_next = line_cost_with_next_word + next_word_cost

        # saven cost with next word taken
        words_cost_guess.append(total_cost_taken_next)
        
        # check if words list traversed
        if next_index < len(words):
            
            # update line lenght with one white space between
            line_len += len(words[next_index]) + 1
            
            if not (line_len <= line_width):
                word_fits = False
        else:
            words_traversed = True

        # update next word index
        next_index += 1

    # minimum cost produces taken from the guesses
    minimum_cost = min(words_cost_guess)
    
    # index of minimum cost next word
    best_next_starting_word_index = word_index + 1 + words_cost_guess.index(minimum_cost)

    # memoizing
    cost_list[word_index] = best_next_starting_word_index
    
    return minimum_cost
        
        
def break_string(string, line_width):
    
    # total words in the string
    words = string.split()
    words_len = len(words)
    
    # list of next word indices producing least cost
    # - index: current word
    # - value: index of the next word that produces the 
    #          least cost at the starting index
    words_least_cost = [None]*words_len
    
    # calculating minimum cost at first index
    initial_word_index = 0
    minimum_cost = break_string_recursive_helper(
        words, line_width, words_least_cost, initial_word_index
    )

    # reconstructing the actual words that produces
    # the maximum space ocupied per line
    words_per_line = []
    word_index_path = 0

    while word_index_path < words_len:

        # next starting line word index
        next_index = words_least_cost[word_index_path]

        # current line words
        word = ' '.join(words[word_index_path:next_index])
        
        # saving word
        words_per_line.append(word)
        
        # updating the index path
        word_index_path = next_index
    
    print('minimum cost: {}'.format(minimum_cost))
    print('list of minimum words cost: {}'.format(words_least_cost))
    print('words breaked: {}'.format(words_per_line))

    return words_per_line
````

````python
string = "the quick brown fox jumps over the lazy dog" 
k = 10
expected_output = ["the quick", "brown fox", "jumps over", "the lazy", "dog"]

assert break_string(string, k) == expected_output
````
