```python
import re
def natural_sort_alphanum(array):
    """ 
    Sort the given list in the way that humans expect.
    """
    convert = lambda text: int(text) if text.isdigit() else text
    alphanum_key = lambda key: [convert(c) for c in re.split('([0-9]+)', key)]
    sorted_array = sorted(array, key=alphanum_key)
    return sorted_array

# natural_sort_alphanum([ 'abl', '1fp']) == ['1fp', 'abl']
```

```python
def prioritize_order_logic(left, right):
    """
    1. The Prime orders should be returned first, sorted by lexicographic sort of the alphabetic metadata.
    2. Only in the case of ties, the identifier should be used as a backup sort.
    3. The remaining non-Prime orders must all come after, in the original order they were given in the input.
    """

    left_code = left.split()[0]
    left_metadata = left.split()[1:]
    left_metadata_index = 0

    right_code = right.split()[0]
    right_metadata = right.split()[1:]
    right_metadata_index = 0

    left_is_prime = True if left_metadata[1].isalpha() else False
    right_is_prime = True if right_metadata[1].isalpha() else False

#     set_trace()
#     print('left_metadata-left_is_prime:\n',left_metadata,left_is_prime)
#     print('right_metadata-right_is_prime:\n', right_metadata, right_is_prime)

    if left_is_prime and not right_is_prime:
        return 'LEFT'

    if not left_is_prime and right_is_prime:
        return 'RIGHT'

    if left_is_prime and right_is_prime:
        while (
            left_metadata_index < len(left_metadata) and
            right_metadata_index < len(right_metadata)
        ):

            left_metadata_element = left_metadata[left_metadata_index]
            right_metadata_element = right_metadata[right_metadata_index]

            if left_metadata_element == right_metadata_element:
                left_metadata_index += 1
                right_metadata_index += 1
                continue

            left_primer_than_right = [
                left_metadata_element, right_metadata_element] == sorted(
                   [left_metadata_element, right_metadata_element]
               )

            if left_primer_than_right: return 'LEFT'
            else: return 'RIGHT'

        # if there is a tie compare against code
        left_primer_than_right = [
                left_code, right_code
            ] == natural_sort_alphanum([left_code, right_code])

        if left_primer_than_right: return 'LEFT'
        else: return 'RIGHT'

    return 'LEFT'
```

```python
def merge(left, right):
    
    sorted_array = []
    left_index, right_index = 0, 0
    left_len, right_len = len(left), len(right)
#     print('initial left and right:\n', left, right)
    while left_index < left_len and right_index < right_len:
        if prioritize_order_logic(left[left_index], right[right_index]) == 'LEFT':
            sorted_array.append(left[left_index])
            left_index += 1
        else:
            sorted_array.append(right[right_index])
            right_index += 1
    
    if left[left_index:]:
        sorted_array.extend(left[left_index:])
    else:
        sorted_array.extend(right[right_index:])
    
#     print('sorted array:\n',sorted_array)
    return sorted_array
    
def prioritize_orders_sort(numOrders, orderList):
    if numOrders <= 1: return orderList
    
    split_index = (numOrders)//2
    left_side = orderList[:split_index]
    right_side = orderList[split_index:]
    
#     print('orderList', orderList)
#     print('left_side', left_side)
#     print('right_side', right_side)
    
    left = prioritize_orders_sort(len(left_side), left_side)
    right = prioritize_orders_sort(len(right_side), right_side)
    
    if [left, right]==[['17g 12 256'], ['Zid 93 12']]:
        set_trace()
    
    return merge(left, right)
```

```python
def prioritizedOrders(numOrders, orderList):
    sorted_list = prioritize_orders_sort(numOrders, orderList)
    return sorted_list
```
